# Question 1.1 (Conditional probability, 10 pts) 

Assume that the conditional probability of an email (chosen uniformly and randomly from a set of emails) containing the word “payment”, given that the email is a spam email, is $72\%$. Suppose that the conditional probability of an email being spam, given that it contains the word “payment”, is $8\%$. Find the ratio of the probability that an email is spam to the probability that an email contains the word “payment”.

**Solution:** 
$$
\begin{aligned}
&\text{P(A) = probability of email containing the word "payment"}\\
&\text{P(B) = probability of email being spam}\\
&\text{P(A|B) = 72\% and P(B|A) = 8\%}\\
&\text{According to Bayes Rule:}\\
&P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}\\
&=\frac{P(A|B)}{P(B|A)} = \frac{P(A)}{P(B)}\\
&=\frac{P(B|A)}{P(A|B)} = \frac{P(B)}{P(A)}\\
&\text{Since we are looking for the ratio of the probability that an email is spam to the probability that an email contains the word payment, we are looking for} \frac{P(B)}{P(A)}\\
&\frac{P(B)}{P(A)} = \frac{P(B|A)}{P(A|B)} = \frac{8\%}{72\%} = 0.111... \text{ or } \frac{1}{9}
\end{aligned}
$$

# Question 1.2 (Conditional probability, 10 pts) 

There are two boxes. Box 1 contains three red and five white balls and box 2 contains two red and five white balls. A box is chosen at random **p(box = 1) = p(box = 2) = 0.5** and a ball chosen at random from this box turns out to be red. What is the posterior probability that the red ball came from box 1?


**Solution:**

Prior: $p(box = 1) = p(box = 2) = 0.5$
$$
\begin{aligned}
\text{Bayes \ Theorem:} \\
&P(box = 1\ |\ R) = \frac{P(R\ |\ box = 1) \cdot P(box = 1)}{P(R)}\\
\\
&P(R\ |\ box = 1) = \frac{3}{8}\\
&P(R\ |\ box = 2) = \frac{2}{7}\\
&P(box = 1) = P(box = 2) = \frac{1}{2}\\

\text{Finding } P(R):\\
&P(R) = P(R\ |\ box = 1) \cdot P(box = 1) + P(R\ |\ box=2) \cdot P(box = 2)\\
&= (\frac{3}{8} \cdot \frac{1}{2}) + (\frac{2}{7} \cdot \frac{1}{2}) = \frac{37}{112}\\
\\
\text{Plugging } P(R) &\text{ back into the equation gives us:} \\
&P(box = 1\ |\ R) = \frac{\frac{3}{8} \cdot \frac{1}{2}}{\frac{37}{112}} = \frac{336}{592} = \frac{21}{37}\approx 0.567 \text{ or } 56.7\%
\end{aligned}
$$

# Question 1.3 (Gaussian & Poisson Distribution, 10 pts) 

Part a) **Gaussian Distribution**

Let $X\sim N(0, 1)$ be a Gaussian random variable, which has the following probability density function:

$$ {f}(x) = \dfrac 1 {\sigma \sqrt{2 \pi} } \exp \bigg({-\dfrac {( {x - \mu})^2} {2 \sigma^2} }\bigg) $$

Find E(X) and show all the derivation steps.

**Solution:**

$$
\begin{aligned}
& X\sim N(0,1) \rightarrow \mu = 0, \sigma = 1\\
& E(X) = \int_{-\infty}^{\infty} x f(x) \,dx \\
& E(X) = \int_{-\infty}^{\infty} x \cdot \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}} dx\\
& E(X) = \int_{-\infty}^{\infty} x \cdot \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} dx\\
& E(X) = \frac{1}{\sqrt{2\pi}} \cdot \int_{-\infty}^{\infty} x \cdot  e^{-\frac{x^2}{2}} dx\\
\text{Since } x \cdot  e^{-\frac{x^2}{2}} \text{ is odd },& E(X) = 0.

\end{aligned}
$$

Part b) **Poisson Distribution**

Let $Y$ be a Poisson random variable, which has the following probability density function:

$$ {f}(y; \lambda) = \Pr(Y{=}k)= \frac{\lambda^k e^{-\lambda}}{k!} $$

Where:
1) **k** is the number of occurences (k=0,1,2...)

2) **e** is Euler's number (e=271828...)

3) **!** is the factorial function

Find E(Y) and show all the derivation steps.

**Solution:**

$$
\begin{aligned}

E(Y) &= \sum_{k=0}^{\infty} k \cdot f(y)\\
E(Y) &= \sum_{k=0}^{\infty} k \cdot \frac{\lambda^k e^{-\lambda}}{k!}\\
&= \sum_{k=0}^{\infty} k \cdot \frac{(\lambda \cdot \lambda^{k-1}) e^{-\lambda}}{k \cdot (k-1)!}\\
&= \lambda e^{-\lambda} \sum_{k=0}^{\infty}\frac{\lambda^{k-1}}{k-1!}\\
\text{Taylor series of } e^{\lambda} &= \sum_{n=0}^{\infty}\frac{\lambda^n}{n!}\\
\text{Therefore, } E(Y) &= \lambda e^{-\lambda} \cdot e^{\lambda} = \lambda

\end{aligned}
$$

# Question 1.4 (Expectation and Variance, 10 pts) 

Suppose that $X_1, ..., X_n$ are independent random variables with the same distribution. 

**(a --5 pts):** Denote the mean of $X_i$ as $\mathbf{E}[X_1]$, find the mean of 
$$\dfrac{X_1+\cdot \cdot \cdot+ X_n}{n}.$$

**(b --5 pts):** Denote the variance of $X_i$ as $\text{Var}[X_1]$, find the variance of 
$$\dfrac{X_1+\cdot \cdot \cdot+ X_n}{n}.$$

**Solution (a)**:

$$\begin{aligned}
\text{mean of } X_i &= E[X_1]\\
\text{mean of } \frac{X_1+...+X_n}{n} &= E[\frac{X_1+...+X_n}{n}]\\
&= \frac{1}{n} \cdot (E[X_1 + X_2 ... + X_n])\\
&= \frac{1}{n} \cdot (E[X_1] + E[X_2] ... + E[X_n])\\
\text{Since } X_i &= E[X_1],\\
&= \frac{1}{n} (n \cdot E[X_1])\\
&= E[X_1]
\end{aligned}$$

**Solution (b)**:

$$\begin{aligned}
\text{Find Var}[\frac{X_1+...+X_n}{n}]\\
&= Var[\frac{1}{n}(X_1+...X_n)]\\
&= (\frac{1}{n})^2 (Var[X_1+...+X_n]) \text{, since } Var[c \cdot X] = c^2 \cdot Var[x]\\
&= (\frac{1}{n})^2 (Var[X_1] + ... + Var[X_n]) \\
&= \frac{1}{n^2} (n \cdot Var[X_1]) \text{, since }X_i = Var[X_1]\\
&= \frac{1}{n} Var[X_1]

\end{aligned}$$
